### 创建 dataframe

https://stackoverflow.com/questions/18837262/convert-python-dict-into-a-dataframe

```python
# dict，默认方式
sales = [{'account': 'Jones', 'jan': 150, 'feb': 200, 'mar': 140},
        {'account': 'Alpha', 'jan': 200, 'feb': 210, 'mar': 215},
        {'account': 'Blue', 'jan': 50, 'feb': 90, 'mar': 95}]
df = pd.DataFrame(sales)
df
```

```python
# from_dict
sales = {'account': ['Jones', 'Alpha', 'Blue'], 
         'jan': [150, 200, 50], 
         'feb': [200, 210, 90], 
         'mar': [140, 215, 95]}
df = pd.DataFrame.from_dict(sales)
df
```

```python
# list，采用函数 from_records
sales = [('Jones', 150, 200, 140),
        ('Alpha', 200, 210, 215),
        ('Blue', 50, 90, 95)]
labels = ['account', 'jan', 'feb', 'mar']
df = pd.DataFrame.from_records(sales, columns=labels)
df
```

```python
# from_items
sales = [('Jones', ['Jones', 'Alpha', 'Blue']),
        ('jan', [150, 200, 50]),
        ('feb', [200, 210, 90]),
        ('mar', [140, 215, 95])]
df = pd.DataFrame.from_items(sales)
df
```
